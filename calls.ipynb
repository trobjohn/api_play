```python
''' Get parameters for make/year/model API data download. '''

import random as rm
import time as tm
import requests
import pickle

## Get all years in which complaints were recorded: minutes
print('Getting years...')
url = 'https://api.nhtsa.gov/products/vehicle/modelYears?issueType=c'
header = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:124.0) Gecko/20100101 Firefox/124.0'} # This is chrome, you can set whatever browser you like
raw = requests.get(url,headers=header)
response_dict = raw.json()
complaint_years = response_dict['results']
years = [ complaint_years[i]['modelYear'] for i in range(len(complaint_years)) ]

## Get all makes for each year in which a complaint occured: minutes
print('Getting makes...')
all_ym = []
url1 = 'https://api.nhtsa.gov/products/vehicle/makes?modelYear='
url2 = '&issueType=c'
for y in years:
    raw = requests.get(url1+y+url2,headers=header)
    response_dict = raw.json()
    all_ym.append(response_dict['results'])

## Get all models for each year: 35 minutes
print('Getting models...')
url1 = 'https://api.nhtsa.gov/products/vehicle/models?modelYear='
url2 = '&make='
url3 = '&issueType=c'
year_make_model = []
for k in range(len(all_ym)):
    this_mm = all_ym[k]
    for j in range(len(this_mm)):
        record = this_mm[j]
        year = record['modelYear']
        make = record['make']
        raw = requests.get(url1+year+url2+make+url3,headers=header)   
        response_dict = raw.json()
        year_make_model.append(response_dict['results'])
    print('    Year: ', year)
    tm.sleep(rm.random() * 3 + 2) # Slow down the request rate

pickle.dump(years, open("./nhtsa_complaints/y.pickle", "wb"))
pickle.dump(all_ym, open("./nhtsa_complaints/ym.pickle", "wb"))
pickle.dump(year_make_model, open("./nhtsa_complaints/ymm.pickle", "wb"))
```
